```python
# Importing necessary libraries

import pandas as pd

df = pd.read_csv(didset.csv)
print(df.head())
```

```python
df = df.filter(['units', 'dpre', 'dpost','chain','buyer_zip'])
```

OLS Estimate for OxyCotin

```python
import statsmodels.api as sm
```

```python
# Define the dependent variable
y = df['units']

# Define the independent variables, adding a constant term for an intercept
X = df[['dpre', 'dpost', 'chain']]
X = sm.add_constant(X)  # adding a constant

# Create the OLS model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the summary of the model
print(results.summary())
```

Export dataframe as .csv for machine learning
